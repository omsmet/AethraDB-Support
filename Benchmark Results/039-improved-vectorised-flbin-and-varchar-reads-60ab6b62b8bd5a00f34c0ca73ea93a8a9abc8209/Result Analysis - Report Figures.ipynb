# AethraDB All Query Result Analysis Overview
This file contains the result analysis for the filter query aggregation and join queries at the current stage of development in the AethraDB engine. Additionally, it contains a comparison against a single-threaded run of the same queries on the same data on DuckDB version 0.8.1 using PyArrow 13.0.0.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('tableau-colorblind10')
plt_patterns = ('-', 'x','/','\\','O','o','//','\\\\')
```

```python
# Load the data from AethraDB
aethra_results = pd.read_csv("./main_method_benchmark_result_OpenJDK-XPS.csv")

aethra_results.head()
```

```python
# Load the data from DuckDB into a single dataframe
duck_files = [
    "duckdb_tpch_q1_query_result_XPS.csv",
    "duckdb_tpch_q3_query_result_XPS.csv",
    "duckdb_tpch_q6_query_result_XPS.csv",
    "duckdb_tpch_q10_query_result_XPS.csv",
    "duckdb_SSB_benchmark_result_XPS.csv"
]
duck_results = {}

for filename in duck_files:
    query_id = "SSB"
    if "SSB" not in filename:
        query_id = filename.replace("duckdb_tpch_q", "").replace("_query_result_XPS.csv", "")
    df = pd.read_csv(filename)
    duck_results[query_id] = df

duck_results["1"].head()
```

# Analysis of TPC-H Queries at different Scale Factors

```python
# Get the data of the query
tpch_data = aethra_results[aethra_results['Benchmark'].str.contains('tpch')].copy(deep=True)

# Introduce numerical TPC-H identifier
tpch_data['TPC-H Query'] = tpch_data['Benchmark'].str.replace("tpch_q", "").astype(int)

# Obtain the different scale factors and queries
sf_vals = tpch_data["Scale-Factor"].unique()
sf_vals.sort()
scale_factors = list(sf_vals)

query_vals = tpch_data["TPC-H Query"].unique()
query_vals.sort()
queries = list(query_vals)

# Plot the data
fig, ax = plt.subplots(nrows=len(query_vals), ncols=len(scale_factors), figsize=(15,20))

for j, query in enumerate(queries):
    for i, sf in enumerate(scale_factors):
        query_data = tpch_data[tpch_data['TPC-H Query'] == query]
        query_sf_data = query_data[query_data['Scale-Factor'] == sf]
        
        duck_query_data = duck_results[str(query)]
        duck_query_sf_data = duck_query_data[duck_query_data['Scale-Factor'] == sf]
        
        duck_data = duck_query_sf_data
        aethra_nv_data = query_sf_data[query_sf_data["Paradigm"] == "non-vectorised"]
        aethra_v_data = query_sf_data[query_sf_data["Paradigm"] == "vectorised"]

        df = pd.DataFrame({
            'DuckDB': duck_data["Running-Time"].tolist(),
            'AethraDB Data-Centric': aethra_nv_data["Total-Time"].tolist(),
            'AethraDB Vectorised': aethra_v_data["Total-Time"].tolist(),
        })

        df.plot.bar(rot=0, ax=ax[j][i])
        if i == 0:
            ax[j][i].set_ylabel("Execution Time (ms)")
        ax[j][i].set_title("TPC-H Q" + str(query) + " Scale-Factor " + str(sf))
        ax[j][i].get_legend().remove()

        bars = ax[j][i].patches
        hatches = [p for p in plt_patterns for j in range(len(df))]
        for bar, hatch in zip(bars, hatches):
            bar.set_hatch(hatch)
        
ax[0][2].legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

# Analysis of SSB Queries at different Scale Factors

```python
def plot_SSB_query_variation(query_variation):
    # Get the data of the query
    SSB_data = aethra_results[aethra_results['Benchmark'].str.contains('ssb_q' + str(query_variation) + '_')].copy(deep=True)

    # Introduce an easier SSB query identifier
    SSB_data['SSB Query'] = SSB_data['Benchmark'].str.replace("ssb_q", "").str.replace("_", ".")

    # Obtain the different scale factors and queries
    sf_vals = SSB_data["Scale-Factor"].unique()
    sf_vals.sort()
    scale_factors = list(sf_vals)

    query_vals = SSB_data["SSB Query"].unique()
    query_vals.sort()
    queries = list(query_vals)

    # Get the SSB duckdb data
    duck_db_SSB = duck_results["SSB"]

    # Plot the data
    vertical_figure_height = 5 * len(queries)
    fig, ax = plt.subplots(nrows=len(query_vals), ncols=len(scale_factors), figsize=(15,vertical_figure_height))

    for j, query in enumerate(queries):
        for i, sf in enumerate(scale_factors):
            query_data = SSB_data[SSB_data['SSB Query'] == query]
            query_sf_data = query_data[query_data['Scale-Factor'] == sf]

            duckdb_query_id = "SSB Q" + query.replace(".", "_")
            duck_query_data = duck_db_SSB[duck_db_SSB["Query"] == duckdb_query_id]
            duck_query_sf_data = duck_query_data[duck_query_data['Scale Factor'] == sf]    

            duck_data = duck_query_sf_data
            aethra_nv_data = query_sf_data[query_sf_data["Paradigm"] == "non-vectorised"]
            aethra_v_data = query_sf_data[query_sf_data["Paradigm"] == "vectorised"]

            df = pd.DataFrame({
                'DuckDB': duck_data["Total Time"].tolist(),
                'AethraDB Data-Centric': aethra_nv_data["Total-Time"].tolist(),
                'AethraDB Vectorised': aethra_v_data["Total-Time"].tolist(),
            })

            df.plot.bar(rot=0, ax=ax[j][i])
            if i == 0:
                ax[j][i].set_ylabel("Execution Time (ms)")
            ax[j][i].set_title("SSB Q" + str(query) + " Scale-Factor " + str(sf))
            ax[j][i].get_legend().remove()

            bars = ax[j][i].patches
            hatches = [p for p in plt_patterns for j in range(len(df))]
            for bar, hatch in zip(bars, hatches):
                bar.set_hatch(hatch)

    ax[0][2].legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

```python
# Plot SSB Q1.*
plot_SSB_query_variation(1)
```

```python
# Plot SSB Q2.*
plot_SSB_query_variation(2)
```

```python
# Plot SSB Q3.*
plot_SSB_query_variation(3)
```
